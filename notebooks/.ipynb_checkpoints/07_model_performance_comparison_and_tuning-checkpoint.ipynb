```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             f1_score, roc_auc_score, roc_curve, confusion_matrix)

df = pd.read_csv('../datasets/processed/data_banjir_processed.csv')

class_names = ['Tidak Banjir', 'Banjir']

X = df.select_dtypes(include=[np.number]).drop(columns=['banjir'], errors='ignore')
y = df['banjir']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

comparison_summary = []

print(f"Data Berhasil Disiapkan dengan Label: {class_names}")
```

```python
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)
y_prob_dt = dt_model.predict_proba(X_test)[:, 1]

fig, ax = plt.subplots(1, 2, figsize=(16, 6))

cm = confusion_matrix(y_test, y_pred_dt)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', ax=ax[0],
            xticklabels=class_names, yticklabels=class_names)
ax[0].set_title('Confusion Matrix: Decision Tree')
ax[0].set_xlabel('Prediksi')
ax[0].set_ylabel('Aktual')

fpr, tpr, _ = roc_curve(y_test, y_prob_dt)
ax[1].plot(fpr, tpr, color='green', lw=3, label=f'AUC = {roc_auc_score(y_test, y_prob_dt):.4f}')
ax[1].plot([0, 1], [0, 1], 'k--', lw=2)
ax[1].set_title('ROC Curve: Decision Tree')
ax[1].legend(loc="lower right")
plt.show()

metrics_dt = {
    'Accuracy': accuracy_score(y_test, y_pred_dt),
    'Precision': precision_score(y_test, y_pred_dt),
    'Recall': recall_score(y_test, y_pred_dt),
    'F1-Score': f1_score(y_test, y_pred_dt),
    'AUC-ROC': roc_auc_score(y_test, y_prob_dt)
}
comparison_summary.append({'Model': 'Decision Tree', **metrics_dt})

print("\n" + "="*40)
print(" MATRIX PERFORMA: DECISION TREE ")
print("="*40)
report_dt = pd.DataFrame(metrics_dt, index=['Score']).T
print(report_dt)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]

fig, ax = plt.subplots(1, 2, figsize=(16, 6))

cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot=True, fmt='d', cmap='Purples', ax=ax[0],
            xticklabels=class_names, yticklabels=class_names)
ax[0].set_title('Confusion Matrix: Random Forest')
ax[0].set_xlabel('Prediksi')
ax[0].set_ylabel('Aktual')

fpr, tpr, _ = roc_curve(y_test, y_prob_rf)
ax[1].plot(fpr, tpr, color='purple', lw=3, label=f'AUC = {roc_auc_score(y_test, y_prob_rf):.4f}')
ax[1].plot([0, 1], [0, 1], 'k--', lw=2)
ax[1].set_title('ROC Curve: Random Forest')
ax[1].legend(loc="lower right")
plt.show()

metrics_rf = {
    'Accuracy': accuracy_score(y_test, y_pred_rf),
    'Precision': precision_score(y_test, y_pred_rf),
    'Recall': recall_score(y_test, y_pred_rf),
    'F1-Score': f1_score(y_test, y_pred_rf),
    'AUC-ROC': roc_auc_score(y_test, y_prob_rf)
}
comparison_summary.append({'Model': 'Random Forest', **metrics_rf})

print("\n" + "="*40)
print(" MATRIX PERFORMA: RANDOM FOREST ")
print("="*40)
report_rf = pd.DataFrame(metrics_rf, index=['Score']).T
print(report_rf)
```

```python
lr_model = LogisticRegression(max_iter=5000, solver='saga', random_state=42)
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)
y_prob_lr = lr_model.predict_proba(X_test)[:, 1]

fig, ax = plt.subplots(1, 2, figsize=(16, 6))

cm = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax[0],
            xticklabels=class_names, yticklabels=class_names)
ax[0].set_title('Confusion Matrix: Logistic Regression')
ax[0].set_xlabel('Prediksi')
ax[0].set_ylabel('Aktual')

fpr, tpr, _ = roc_curve(y_test, y_prob_lr)
ax[1].plot(fpr, tpr, color='blue', lw=3, label=f'AUC = {roc_auc_score(y_test, y_prob_lr):.4f}')
ax[1].plot([0, 1], [0, 1], 'k--', lw=2)
ax[1].set_title('ROC Curve: Logistic Regression')
ax[1].legend(loc="lower right")
plt.show()

metrics_lr = {
    'Accuracy': accuracy_score(y_test, y_pred_lr),
    'Precision': precision_score(y_test, y_pred_lr),
    'Recall': recall_score(y_test, y_pred_lr),
    'F1-Score': f1_score(y_test, y_pred_lr),
    'AUC-ROC': roc_auc_score(y_test, y_prob_lr)
}
comparison_summary.append({'Model': 'Logistic Regression', **metrics_lr})

print("\n" + "="*40)
print(" MATRIX PERFORMA: LOGISTIC REGRESSION ")
print("="*40)
report_lr = pd.DataFrame(metrics_lr, index=['Score']).T
print(report_lr)
```

```python
df_final = pd.DataFrame(comparison_summary).set_index('Model')

print("\n" + "="*80)
print(" RINGKASAN AKHIR KOMPARASI SEMUA MODEL ")
print("="*80)

display(df_final.sort_values(by='F1-Score', ascending=False).style.background_gradient(cmap='YlGnBu'))

df_final.plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title('Perbandingan Metrik Semua Model (Banjir vs Tidak Banjir)')
plt.ylabel('Skor (0-1)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

param_grid_dt = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_dt, cv=5, scoring='f1')
grid_dt.fit(X_train, y_train)

best_dt = grid_dt.best_estimator_
y_pred_dt_tuned = best_dt.predict(X_test)
y_prob_dt_tuned = best_dt.predict_proba(X_test)[:, 1]

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
sns.heatmap(confusion_matrix(y_test, y_pred_dt_tuned), annot=True, fmt='d', cmap='Greens', ax=ax[0],
            xticklabels=class_names, yticklabels=class_names)
ax[0].set_title('Tuned Decision Tree: Confusion Matrix')

fpr, tpr, _ = roc_curve(y_test, y_prob_dt_tuned)
ax[1].plot(fpr, tpr, color='green', lw=3, label=f'AUC = {roc_auc_score(y_test, y_prob_dt_tuned):.4f}')
ax[1].plot([0, 1], [0, 1], 'k--')
ax[1].legend()
plt.show()

metrics_dt_tuned = {
    'Model': 'Decision Tree (Tuned)',
    'Accuracy': accuracy_score(y_test, y_pred_dt_tuned),
    'Precision': precision_score(y_test, y_pred_dt_tuned),
    'Recall': recall_score(y_test, y_pred_dt_tuned),
    'F1-Score': f1_score(y_test, y_pred_dt_tuned),
    'AUC-ROC': roc_auc_score(y_test, y_prob_dt_tuned)
}
comparison_summary.append(metrics_dt_tuned)
print(f"Best Parameters: {grid_dt.best_params_}")
display(pd.DataFrame(metrics_dt_tuned, index=[0]).T)
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 15, 25],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

random_rf = RandomizedSearchCV(RandomForestClassifier(random_state=42), param_dist_rf, 
                               n_iter=10, cv=3, scoring='f1', random_state=42)
random_rf.fit(X_train, y_train)

best_rf = random_rf.best_estimator_
y_pred_rf_tuned = best_rf.predict(X_test)
y_prob_rf_tuned = best_rf.predict_proba(X_test)[:, 1]

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
sns.heatmap(confusion_matrix(y_test, y_pred_rf_tuned), annot=True, fmt='d', cmap='Purples', ax=ax[0],
            xticklabels=class_names, yticklabels=class_names)
ax[0].set_title('Tuned Random Forest: Confusion Matrix')

fpr, tpr, _ = roc_curve(y_test, y_prob_rf_tuned)
ax[1].plot(fpr, tpr, color='purple', lw=3, label=f'Tuned AUC = {roc_auc_score(y_test, y_prob_rf_tuned):.4f}')
ax[1].plot([0, 1], [0, 1], 'k--')
ax[1].legend()
plt.show()

metrics_rf_tuned = {
    'Model': 'Random Forest (Tuned)',
    'Accuracy': accuracy_score(y_test, y_pred_rf_tuned),
    'Precision': precision_score(y_test, y_pred_rf_tuned), # <--- SEKARANG ADA
    'Recall': recall_score(y_test, y_pred_rf_tuned),
    'F1-Score': f1_score(y_test, y_pred_rf_tuned),
    'AUC-ROC': roc_auc_score(y_test, y_prob_rf_tuned)
}
comparison_summary.append(metrics_rf_tuned)
print(f"Best Parameters: {random_rf.best_params_}")
display(pd.DataFrame(metrics_rf_tuned, index=[0]).T)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'solver': ['saga'] # SAGA mendukung l1 dan l2
}

grid_lr = GridSearchCV(LogisticRegression(max_iter=5000, random_state=42), param_grid_lr, cv=5, scoring='f1')
grid_lr.fit(X_train, y_train)

best_lr = grid_lr.best_estimator_
y_pred_lr_tuned = best_lr.predict(X_test)
y_prob_lr_tuned = best_lr.predict_proba(X_test)[:, 1]

print(f"Best Parameters: {grid_lr.best_params_}")

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
sns.heatmap(confusion_matrix(y_test, y_pred_lr_tuned), annot=True, fmt='d', cmap='Blues', ax=ax[0],
            xticklabels=class_names, yticklabels=class_names)
ax[0].set_title('Tuned Logistic Regression: Confusion Matrix')

fpr, tpr, _ = roc_curve(y_test, y_prob_lr_tuned)
ax[1].plot(fpr, tpr, color='blue', lw=3, label=f'Tuned AUC = {roc_auc_score(y_test, y_prob_lr_tuned):.4f}')
ax[1].plot([0, 1], [0, 1], 'k--')
ax[1].legend()
plt.show()

metrics_lr_tuned = {
    'Model': 'Logistic Regression (Tuned)',
    'Accuracy': accuracy_score(y_test, y_pred_lr_tuned),
    'Precision': precision_score(y_test, y_pred_lr_tuned), # <--- SEKARANG ADA
    'Recall': recall_score(y_test, y_pred_lr_tuned),
    'F1-Score': f1_score(y_test, y_pred_lr_tuned),
    'AUC-ROC': roc_auc_score(y_test, y_prob_lr_tuned)
}
comparison_summary.append(metrics_lr_tuned)
display(pd.DataFrame(metrics_lr_tuned, index=[0]).T)
```

```python
df_all = pd.DataFrame(comparison_summary)
df_all['Version'] = df_all['Model'].apply(lambda x: 'Tuned' if 'Tuned' in x else 'Base')
df_all['Algorithm'] = df_all['Model'].str.replace(' (Tuned)', '', regex=False)

# Mengurutkan agar Base dan Tuned muncul berurutan per algoritma
df_report = df_all.sort_values(by=['Algorithm', 'Version'], ascending=[True, False])

print("\n" + "="*80)
print("             TABEL EVALUASI KOMPARATIF (BASE VS TUNED)")
print("="*80)

styled_report = df_report.drop(columns=['Algorithm', 'Version']).set_index('Model').style \
    .background_gradient(cmap='Blues', subset=['Accuracy', 'Precision', 'Recall', 'F1-Score', 'AUC-ROC']) \
    .highlight_max(color='lightgreen', axis=0) \
    .format("{:.4f}")

display(styled_report)
```

```python
metrics = ['F1-Score', 'AUC-ROC', 'Recall', 'Precision']
df_melt = df_all.melt(id_vars=['Algorithm', 'Version'], value_vars=metrics, 
                      var_name='Metric', value_name='Score')

min_score = df_melt['Score'].min()
y_limit_lower = max(0, float(f"{min_score:.1f}") - 0.1) # Contoh: jika min 0.82, mulai dari 0.7

plt.figure(figsize=(16, 8))
sns.set_style("whitegrid")

palette = {"Base": "#bdc3c7", "Tuned": "#2ecc71"}
ax = sns.barplot(data=df_melt, x='Metric', y='Score', hue='Version', 
                 palette=palette, edgecolor='0.2', linewidth=1.5)

plt.ylim(y_limit_lower, 1.05) 

for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{height:.3f}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='baseline', fontsize=11, fontweight='bold',
                    xytext=(0, 10), textcoords='offset points', color='#2c3e50')

plt.title('Performance Jump: Base vs Tuned Models (Zoomed View)', fontsize=18, fontweight='bold', pad=25)
plt.ylabel(f'Performance Score (Starting from {y_limit_lower})', fontweight='bold')
plt.xlabel('Evaluation Metrics', fontweight='bold')
plt.legend(title='Version', loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2, shadow=True)

sns.despine()
plt.tight_layout()
plt.show()
```

```python
df_pivot = df_all.pivot(index='Algorithm', columns='Version', values='F1-Score')
df_pivot['Improvement (%)'] = ((df_pivot['Tuned'] - df_pivot['Base']) / df_pivot['Base']) * 100

print("\n" + "="*80)
print("             ANALISIS PERSENTASE PENINGKATAN F1-SCORE")
print("="*80)
display(df_pivot.style.format("{:+.2f}%", subset=['Improvement (%)'])
        .background_gradient(cmap='RdYlGn', subset=['Improvement (%)']))
```

```python
feature_names = X.columns

def get_importance_df(model, name):
    importances = model.feature_importances_
    return pd.DataFrame({
        'Feature': feature_names,
        'Importance': importances,
        'Model': name
    }).sort_values(by='Importance', ascending=False)

def get_coef_df(model, name):
    coefs = np.abs(model.coef_[0]) 
    return pd.DataFrame({
        'Feature': feature_names,
        'Importance': coefs,
        'Model': name
    }).sort_values(by='Importance', ascending=False)

df_rf_imp = get_importance_df(best_rf, 'Random Forest (Tuned)')
df_dt_imp = get_importance_df(best_dt, 'Decision Tree (Tuned)')
df_lr_imp = get_coef_df(best_lr, 'Logistic Regression (Tuned)')

fig, axes = plt.subplots(3, 1, figsize=(12, 18))

models_data = [
    (df_rf_imp, 'Random Forest (Tuned) - Feature Importance', 'Purples_r'),
    (df_dt_imp, 'Decision Tree (Tuned) - Feature Importance', 'Greens_r'),
    (df_lr_imp, 'Logistic Regression (Tuned) - Absolute Coefficients', 'Blues_r')
]

for i, (df_imp, title, color) in enumerate(models_data):
    # Perbaikan di sini: Menambahkan hue dan legend=False untuk menghindari FutureWarning
    sns.barplot(data=df_imp.head(10), x='Importance', y='Feature', 
                hue='Feature', palette=color, ax=axes[i], edgecolor='0.3', legend=False)
    
    axes[i].set_title(title, fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Importance Score / Absolute Coef')
    axes[i].set_ylabel('Features')

plt.tight_layout()
plt.show()

print("\n" + "="*50)
print("TOP 5 FEATURES PER MODEL (TUNED)")
print("="*50)
for df_imp, title, _ in models_data:
    print(f"\nModel: {title}")
    print(df_imp.head(5)[['Feature', 'Importance']].to_string(index=False))
```
